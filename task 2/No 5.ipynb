```python
import sklearn, sklearn.datasets
digits = sklearn.datasets.load_digits()
```

```python
A = digits.data
y = digits.target
```

```python
from matplotlib import pyplot as plt
%config InlineBackend.figure_format = 'svg'
```

```python
fig, axs = plt.subplots(2, 5, figsize = (10, 4))
for i, ax in enumerate(axs.flatten()):
    ax.imshow(A[i].reshape(8, 8), cmap = 'Greys_r')
```

В задании сказано отнормировать датасет. Это можно сделать несколькими способами:

a) нормировать каждое изображение на единицу

b) нормировать каждую фичу

c) Нормировать каждое изображение на среднюю норму

Выберем способ b).

```python
A_normalized = sklearn.preprocessing.scale(A)
```

Метод анализа главных компонент:

```python
U, S, _ = sklearn.utils.extmath.randomized_svd(A_normalized, 2, random_state = 0)
A_reduced = U * S
```

```python
p = plt.scatter(
    x = A_reduced[:, 0],
    y = A_reduced[:, 1],
    c = y,
)
plt.gca().legend(*p.legend_elements())
```
